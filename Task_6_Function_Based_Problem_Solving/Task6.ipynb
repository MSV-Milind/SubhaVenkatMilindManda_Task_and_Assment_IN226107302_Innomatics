## Problem Statement 1: Smart Parking Lot Management System

Create funtion to manage smart parking lot

```python
def smart_total(capacity, logs):
    # Variables to keep track of counts
    entry_count = 0
    exit_count = 0
    peak = 0
    
    # Calculate counts for entry and exit
    for i in logs:
        if i == "IN":
            entry_count += 1
        elif i == "OUT":
            exit_count += 1
        
        current = entry_count - exit_count

        # Check for peak usage
        if current < 0:
            current = 0
        if current > peak:
            peak = current

    # Calculate total parked vehicles
    if entry_count > exit_count:
        total = entry_count - exit_count
    else:
        total = 0

    # Check for parking status
    if total  > capacity:
        status = "Unavailable"
    else:
        status = "Available"
    
    return total, status, peak

# main
ParkingCapacity = 50
VehicleLogs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]
currently_parked, parking_status, peak = smart_total(ParkingCapacity, VehicleLogs)
print(f"Currently Parked Vehicles: {currently_parked}")
print(f"Parking Status: {parking_status}")
# print(f"Peak Usage: {peak}")
```

Technically "OUT" count of vehicles shouldn't be exceeding the "IN" count of vehicles, if it does then we just get the total parked vehicles as 0. Since, at that point there should be no vehicles present in the parking lot

## Problem Statement 2: Online Food Delivery Time Estimator

Create function to estimate delivery time based on distance, weather and traffic

```python
def delivery_estimator(distance, weather, traffic):
    # Setting a base time of 1.5 minutes per km
    base_time = 1.5

    # Calculating time for given distance
    distance_time = distance * base_time

    # Setting weather and traffic delay time variables
    weather_delay = 0
    traffic_delay = 0

    # Check for weather conditions, set delayed time
    if weather == "Rainy":
        weather_delay = 20
    elif weather == "Windy":
        weather_delay = 15
    
    # Check for traffic condtions, set delayed time
    if traffic == "Light":
        traffic_delay = 2
    elif traffic == "Moderate":
        traffic_delay = 12
    elif traffic == "High":
        traffic_delay = 23

    # Calculate estimated time
    estimated_time = distance_time + weather_delay + traffic_delay

    return estimated_time

# main
Distance_km = 8
TrafficLevel = "High"
Weather = "Rainy"
res = delivery_estimator(Distance_km, Weather, TrafficLevel)
print(f"Estimated Delivery Time: {res:.0f} minutes")
```

## Problem Statement 3: Movie Theatre Seat Occupancy Analyzer

Create function to analyze seat booking data

```python
def occupancy_analyzer(seats, booked):
    # Calculate occupancy percentage
    total = 0
    for i in booked:
        total += i
    occupancy = (total / seats) * 100

    # Check for theatre oocupancy status
    if occupancy == 100:
        status = "Housefull"
    elif occupancy >= 95:
        status = "Almost Housefull"
    elif occupancy >= 75:
        status = "Almost Full"
    elif occupancy >= 50:
        status = "Filling Fast"
    else:
        status = "Seats Available"
    
    # Suggest additional shows if housefull
    if status == "Housefull":
        print("Additional shows required")
    
    return occupancy, status

# main
TotalSeats = 200
BookedSeats = [1] * 150
occupancy_percentage, show_status = occupancy_analyzer(TotalSeats, BookedSeats)
print(f"Occupancy: {occupancy_percentage:.0f}%")
print(f"Show Status: {show_status}")
```

##  Problem Statement 4: Cloud Server Load Classification System

Create function to classify server load

```python
def load_classification(cpu_usage):
    # Calculate average 
    total = 0
    for i in cpu_usage:
        total += i
    average = (total / len(cpu_usage)) # (the readings are already in %)

    # Check for average
    if average > 80:
        status = "Critical"
    elif average >= 50:
        status = "Warning"
    else:
        status = "Normal"

    return average, status

# main
CPUReadings = [45, 60, 70, 85, 90]
average_laod, server_status = load_classification(CPUReadings)
print(f"Average CPU Load: {average_laod:.0f}%")
print(f"Server Status: {server_status}")
```

##  Problem Statement 5: Smart Classroom Resource Usage Monitor

Create function to identify overuse patterns for classroom resources

```python
def resource_monitor(usage):

    # Create empty list
    overused = []

    # Check for overusage of resource
    for resource in usage:
        if usage[resource] > 8:
            overused.append(resource)
    
    # Condition for alert status
    if overused:
        alert = "Yes"
    else:
        overused.append("None")  # if no overuse, just gives None
        alert = "No"
    
    return ", ".join(overused), alert

# main
ResourceUsage = {
    "Projector" : 6,
    "AC" : 9,
    "Lights" : 4
}
overused_resources, energy_alert = resource_monitor(ResourceUsage)
print(f"Overused Rsources: {overused_resources}")
print(f"Energy Alert: {energy_alert}")
```

##  Problem Statement 6: Online Event Registration Capacity Controller

Create function to manage event registration

```python
def event_manager(capacity, registrations):
    confirmed = 0
    waitlisted = 0

    # count confirmed and waitlisted users
    for i in range(registrations):
        if confirmed < capacity:
            confirmed += 1
        else:
            waitlisted += 1

    # Condition to update status
    if confirmed < capacity :
        status = "Open"
    else:
        status = "Closed"

    return confirmed, waitlisted, status

# main
EventCapacity = 100
Registrations = 105
confirmed_registrations, waitlisted_users, registration_status = event_manager(EventCapacity, Registrations)
print(f"Confirmed Registrations: {confirmed_registrations}")
print(f"Waitlisted Users: {waitlisted_users}")
print(f"Registration Status: {registration_status}")
```
