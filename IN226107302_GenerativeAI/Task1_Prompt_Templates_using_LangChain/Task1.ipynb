### Task 5: Input Validation Layer 

```python
def validate_inputs(audience, tone):
    sample_audience = ["beginner", "intermediate", "expert"]
    sample_tone = ["formal", "casual", "fun"]
    default_audience = "expert"
    default_tone = "formal"

    if audience not in sample_audience:
        print("Invalid audience! Using Default audience now!")
        audience = default_audience
    if tone not in sample_tone:
        print("Invalid tone! Using Default tone now!")
        tone = default_tone
    
    return audience, tone

print("-" * 35)
input_audience, input_tone = validate_inputs("advanced", "casual")
print(f"Audience: {input_audience}")
print(f"tone: {input_tone}")

print("-" * 35)
input_audience, input_tone = validate_inputs("expert", "unfriendly")
print(f"Audience: {input_audience}")
print(f"tone: {input_tone}")

print("-" * 35)
input_audience, input_tone = validate_inputs("beginner", "fun")
print(f"Audience: {input_audience}")
print(f"tone: {input_tone}")
```

### Task 6: Prompt Generator App

```python
def generate_prompt(topic, audience, tone, style):
    audience,tone = validate_inputs(audience,tone)
    styles = {
    "teaching":"Explain {topic} for a {audience} in a {tone} teaching style",
    "interview":"Ask a few interview questions about {topic} to a {audience} in a {tone} style",
    "storytelling":"Explain {topic} for a {audience} in a {tone} storytelling style"
    }

    default_style = "teaching"

    if style not in styles:
        style = default_style

    template = styles[style]
    prompt = PromptTemplate(
        input_variables=["topic","audience","tone"],
        template=template
    )

    return prompt.format(topic=topic, audience=audience, tone=tone)

print(generate_prompt("Neural Networks","beginner","fun","storytelling"))
```

### Task 7: Template Reusability Test

```python
inputs = [
    {"topic": "LangChain", "audience": "expert", "tone": "formal", "style": "teaching"},
    {"topic": "Python", "audience": "beginner", "tone": "fun", "style": "storytelling"},
    {"topic": "NLP", "audience": "intermediate", "tone": "non-friendly", "style": "teaching"},
    {"topic": "Data Science", "audience": "analysts", "tone": "casual", "style": "interview"},
    {"topic": "Deep Learning", "audience": "engineers", "tone": "technical", "style": "interview"}
]

for i in inputs:
    print("-" * 35)
    print(generate_prompt(**i))
```
