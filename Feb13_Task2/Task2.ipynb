## Problem 1: Unique Words in a Sentence

Given sentence, identify unique words and print count and the words themselves

#### 1.1 Simple method using set

```python
sentence = "python is easy and python is powerful"

# Splitting
words = sentence.split()

# Removing duplicates
unique_words = set(words)

# Printing
print("Unique words count: ", unique_words)
print("Unique words: ", unique_words)
```

#### 1.2 Using dictionary

```python
# Using dictionary helps in preserving the order and removes any duplicates
sentence = "python is easy and python is powerful"

res = {}

list_sentence = sentence.split()

for i in list_sentence:
    res[i] = res.get(i,0) + 1
print("Unique words count: ", len(res)) 
print("Unique words: ", list(res))    # note that the output is in list
```

## Problem 2: Highest Salary from Employee Data

Given a dictionary of employee names and slaries, figure out the highest salary

```python
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

# Using variables to store empty/default values
max_salary = 0
max_name = ""

for employee in employees:
    # Compare each employee salary with the highest salary
    if employees[employee] > max_salary:
        max_salary = employees[employee]    # Update salary and name
        max_name = employee
print(f"Highest Salary: {max_name} - {max_salary}")
```

## Problem 3: Find Maximum and Minimum Values

```python
numbers = [45, 22, 89, 10, 66]

# use two variables to store the first element as max and min
current_max = numbers[0]
current_min = numbers[0]

# Compare each element other than first one with the subsequent elements
for i in range(1, len(numbers)):
    if numbers[i] > current_max:
        current_max = numbers[i]

    if numbers[i] < current_min:
        current_min = numbers[i]

print(f"Max: {current_max}")
print(f"Min: {current_min}")
```

## Problem 4: Count Products Above a Price Threshold

Given, list of integers, find out the count of integers higher than 1000

```python
prices = [450, 1200, 899, 1500, 300]
count = 0

# Compare each element with 1000 and count how many are greater than 1000
for price in prices:
    if price > 1000:
        count += 1

print(f"Products above 1000: {count}")
```

## Problem 5: Calculate Attendance Percentage

Given, list of present and absents, calculate percentage of present

```python
attendance = ["P", "P", "A", "P", "P"]
count = 0

# Check for "P", update counter
for i in attendance:
    if i == "P":
        count += 1
        
# Calculate percentage
present_percentage = (count/len(attendance)) * 100

print(f"Attendance Percentage: {present_percentage}")
```

## Problem 7: Count Character Frequency

Given, a string, count the amount of times each character appears

```python
string = "ppython"
res = {}

# Increment count for each occurrence
for s in string:
    if s in res:
        res[s] += 1
    else:
        res[s] = 1
print(res)
```

```python
# Using dict.get function
string = "ppython"

res = {}

for i in string:
    res[i] = res.get(i, 0) + 1

print(res)
```

We use dict.get() function to get the value of a specific key, if there is no value then python assigns 0 to that key
