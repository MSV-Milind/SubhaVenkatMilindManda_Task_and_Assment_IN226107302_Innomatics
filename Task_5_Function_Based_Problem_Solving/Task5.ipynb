## Problem 1: Social Media - Post Engagement Analyzer

Create fuction to calculate social media engagement of users

```python
# function to calculate engagement
def user_engagement(likes):
    total_likes = 0
    post_type = "Normal Engagement"

    # Sum of likes
    for i in likes:
        total_likes += i

    # If condition
    if total_likes >= 1000:
        post_type = "Viral Post"
        
    print(f"Total Likes: {total_likes}")
    print(f"Post Status: {post_type}")

# main 
user_likes = [100, 200, 300, 450]
user_engagement(user_likes)
```

## Problem 3: Agriculture - Rainfall Adequacy Checker

Create function to analyze seasonal ranifall data

```python
# function to analyze rainfall level
def rainfall_checker(rainfall):
    total_rainfall = 0

    # Calculate total rainfall
    for i in rainfall:
        total_rainfall += i

    # Calculate average
    average = total_rainfall / len(rainfall)

    # Check for rainfall level
    if average >= 65:
        print(f"Average Rainfall: {average:.0f}")
        print("Rainfall Status: Adequate Rainfall")
    else:
        print(f"Average Rainfall: {average:.0f}")
        print("Rainfall Status: Inadequate Rainfall")

# main
rainfall = [55, 65, 98, 70, 77, 68, 71]
rainfall_checker(rainfall)
```

## Problem 4: Social Media - Dupilcate Account Detection

Create function to detect duplicate usernames in system

```python
# function to detect duplicates
def duplicate_detection(usernames):
    if len(usernames) != len(set(usernames)):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")

# main
usernames = ["alex", "john", "jane", "sarah", "john"]
duplicate_detection(usernames)
```

## Problem 6: Agriculture - Premium Crop Price Filter

Create function to filter crop prices based on a threshold

```python
# function to find types of crops
def crop_filter(crop_prices):
    # making two lists to get both premium and regular crops
    premium_crop = []
    regular_crop = []

    # Check for premium crops
    for price in crop_prices:
        if price > 2000:
            premium_crop.append(price)
        else:
            regular_crop.append(price)
    return premium_crop, regular_crop

# main
crop_prices = [1000, 1750, 2500, 1999, 3200, 2000]
premium, regular = crop_filter(crop_prices)
print(f"Premium Crops: {premium}")
```

## Problem 7: System Monitoring - Application Health Checker

Create function to evaluate application health based on error count

```python
# function to evaluate application health
def error_checker(error_count):
    if error_count == 0:
        return "Healthy"
    elif error_count <= 5:
        return "Minor Issues"
    else:
        return "Critical Issues"

# main
error_count = 7
health_status = error_checker(error_count)
print(f"Error Count: {error_count}")
print(f"System Status: {health_status}")
```

## Problem 8: Banking - Daily Transaction Limit Checker

Create function to validate daily transaction amount

```python
# function to validate transaction
def transaction_validation(amount):
    daily_limit = 50000

    if amount > daily_limit:
        status = "Rejected"
    else:
        status = "Approved"

    return status

# main
amount = 40000
transaction_status = transaction_validation(amount)
print(f"Transaction Amount: {amount}")
print(f"Transaction Status: {transaction_status}")
```

## Problem 9: E-Learning - Student Attendance Eligibility System

Create function to calculate attendance percentage and check if the student is eligible for exam based on it.

```python
# function to calculate attendance percentage and exam eligibility
def exam_eligibility(attendance):
    total = 0

    # Calculate total
    for i in attendance:
        total += i

    # Calculate average
    average = total / len(attendance)

    # Check for exam eligibility
    if average >= 75:
        print(f"Attendance Percentage: {average}")
        print("Exam Eligibility: Eligible")
    else:
        print(f"Attendance Percentage: {average}")
        print("Exam Eligibility: Not Eligible")

# main
attendance = [75, 90, 100, 55, 80]
exam_eligibility(attendance)
```

## Problem 10: Smart Electricity Bill Analyzer

Create function to calculate bill usage and classifies based on units consumed

```python
# function to calculate bill and classify bill usage
def billing_system(units):

    # calculate total cost
    if units <=100:
        cost = 100 * 3
    elif units <= 200:
        cost = (100 * 3) + (units - 100) * 5
    else:
        cost = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    # Identify usage
    if cost < 500:
        status = "Low Usage"
    elif cost > 500 and cost < 1500:
        status = "Moderate Usage"
    else:
        status = "High Usage"

    return cost, status

# main 
units = 275
bill, usage_status = billing_system(units)
print(f"Total bill is: {bill}")
print(f"Your usage is {usage_status}")
```
