## Problem 1: Mobile Recharge Validation System

Create function to validate mobile recharge request

```python
def mobile_recharge():
    valid_plans = [199, 299, 399, 599]

    # Runs a while loop till correct amount is selected
    while True:
        amount = int(input("Enter a recharge amount: "))
        if amount < 50:
            print("Try again")
        elif amount not in valid_plans:
            print("Please select a valid plan of 199, 299, 399 or 599")
            print("Try again")

        # Valid rechrage
        else:
            print("Recharge Successful")
            break
            
mobile_recharge()
```

## Problem 2: Inventory Reorder Alert System

Create function to check for stock and reorder if necessary

```python
# dictionary of items and their quantities
warehouse = {
    "Books":100,
    "Mobiles":5,
    "Electronic Accessories":70,
    "Medical devices":14,
    "Paper Supplies":70,
    "Games":15
}


def check_stock():
    print("--------------------------------------")
    print("   Status   |   Item   |   Quantity   ")
    print("--------------------------------------")

    # Checks item quantities
    for stock in warehouse:
        if warehouse[stock] < 15:
            print(f"Reorder Alert → {stock} : {warehouse[stock]}")
        else:
            print(f"Stock OK      → {stock} : {warehouse[stock]}")

check_stock()
```

## Problem 3: Student Result Processing System

Create funtion to process Pass/Fail for students based on average marks

```python
# Simple function with list already given
def student_results(marks):
    total = 0
    for i in marks:
        total += i
    average = total/len(marks)
    if average >= 50:
        return "Pass"
    else:
        return "Fail"

# main
student = [17, 99, 57, 78, 63]
res = student_results(student)
print(res)
```

## Problem 4: Cab Fare Estimator with Retry Option

Create function with fare calculator and retry option

```python
def fare_estimator(total_km, hour_type):
    base_fare = 50
    cost_per_km = 12

    # Peak hour cost calculation
    cost = base_fare + (cost_per_km * total_km)

    # Check for 'peak hour'
    if hour_type == "peak hour":
        cost = cost * 1.25
        
    return cost

# main
while True:
    print("-----Fare Calculator-----")
    print("1.Calculate fare")
    print("2. Exit")
    user_input = int(input("Enter a choice: "))
    if user_input == 1:
                    
        total_km = float(input("Enter total km travelled: "))
        hour_type = input("Enter 'regular hour' or 'peak hour': ")

        # check for positive km distance
        if total_km < 1:
            print("Enter positive km!\n")
            continue

        # Get total fare
        res = fare_estimator(total_km, hour_type)
        print(f"\nYour fare is: {res}\n")

    # Exit condition
    if user_input == 2:
        print("Exiting successfully")
        break
```

## Problem 5: Employee Attendance Eligibility Checker

Create function to check employee eligibilty based on attendance

```python
def eligibility_checker(attendance):

    # calculate total present days
    present_days = 0
    for i in attendance:
        if i == "P":
            present_days += 1

    # calculate average percent
    present_percentage = (present_days/len(attendance)) * 100

    # Condition for eligibility
    if present_percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"

# main
attendance = list(input("Enter 'P' for Present and 'A' for Absent: ").split())
res = eligibility_checker(attendance)
print(res)
```

## Problem 6: Password Strength Checker

Crete function to check password strength based on rules

```python
def password_strength(password):
    # check for character length
    if len(password) < 8:
        return "Password must be atleast 8 characters"
        
    else:
        digit_present = False
        special = False

        # check digits and special characters
        for i in password:
            if i.isdigit():
                digit_present = True
            if i == "@" or i == "#" or i == "$":
                special = True

            # return strong password
            if digit_present and special:
                return "Passowrd is strong"
                
        # Conditions for when one or more rule fails
        if not digit_present:
            print("Password must contain atleast one digit")
        if not special:
            print("Password must contain atleast one special character '@#$'")
        return "Password is weak"

# main
password = input("Enter password: ")
res = password_strength(password)
print(res)
```
