## Q1. User Login Check

```python
username = "admin"
password = "1234"

user_input = input("Enter Username: ")
password_input = input("Enter Password: ")

if user_input == username and password_input == password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

```python
# Checking for the else condition as well
user_input = input("Enter Username: ")
password_input = input("Enter Password: ")

if user_input == username and password_input == password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

So, we are given username and password and are required to check whether any user trying to login is successful or not based on our stored values.
We use a simple if else loop to track successful or failed login.

## Q2. Pass/Fail Analyzer

Given a list of students marks, analyze students who passed or failed and get the total count

```python
marks = [45, 78, 90, 33, 60]
passed = 0
failed = 0

for i in range(len(marks)):
   if marks[i] >= 50: 
       passed += 1
   else:
       failed += 1
print("The total number of pass students is: ", passed)
print("The total number of fail students is: ", failed)
```

We use a simple for loop and an if else statement to go over our list of students marks and evaluate whether passsed or failed. We use two variables to keep track of the count of students who passed and failed.

## Q3. Simple Data Cleaner

Given a list of names, standardize them

```python
names = [" Alice ", "bob", " CHARLIE "]
res = []

for name in names:
    res.append(name.strip().lower())
print("The standardized user names are: ",res)
```

#### Let's walk through what's happening:
res is empty list,

a for loop is used,

first, each name/element in the list is stripped of any whitespaces,

then lower function does a lowercase of all the names,

then we append it to a final list,

print out the list

## Q4. Message Length Analyzer

Given a list of messages, print length of each and flag messages greater than 10 characters

```python
messages = ["Hi", "Welcome to the platform", "OK"]

for message in messages:
    print(f"The length of '{message}' is {len(message)}")
    if len(message) > 10:
        print(f"\nFLAG:'{message}' is longer than 10 characters\n")   # \n is used to create white space
```

```python
# adding an extra for loop to seperate length and flagging messages for readability
messages = ["Hi", "Welcome to the platform", "OK"]

for message in messages:
    print(f"The length of '{message}' is {len(message)}")

for message in messages:
    if len(message) > 10:
        print(f"\nFLAG:'{message}' is longer than 10 characters")
```

The length of each message is printed and any length which is greater than 10 is raised as a flag.

## Q5. Error Message Detector

Given a list of logs, count total errors and print total error count

```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
error_count = 0

for log in logs:
    if log == "ERROR":
        error_count += 1
print("The total 'ERROR' count is: ", error_count)
```

We use a for loop to go throgh our logs and a count variable to keep track of total errors.
