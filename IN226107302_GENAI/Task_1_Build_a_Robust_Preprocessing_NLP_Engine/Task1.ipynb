#### Q1. What is the difference between "Love" and "love" in NLP?

The "Love" is a capitalized word with "L" and "love" is an all lowercase word. Even though it reads like the same word for a human when it comes to machines and in NLP, each word is characterized differently. It sees that the ASCII value of "L" is different from that of "l" and when comparing both words is unable to match them.

#### Q2. What happens if stopwords are not removed?

Stopwords are extremely useful when it comes to human speech as it helps us recognize the intentions and nuances of the speaker. 

However, in some NLP tasks, the machine doesn't need to know the full context. In that case, the more the stopwords, the more computational load on the machine and also increased noise in the data. And stopwords end up having high frequency count which gives a wrong or unneeded conclusion to our analysis. 

#### Q3. Provide two real-world scenarios where removing stopwords can be harmful.

1) For things like translating languages. Stopwords would actually help to accurately translate the literal meaning of a foreign language.

2) For Sentiment Analysis. Stopwords like no, not can completely change the meaning of a sentence. So, if stopwords are removed, say from customer reviews, then the sentence can take an unintended meaning and create inherent bias in our analysis.

#### Q4. Explain the difference between stemming and lemmatization.


Stemming uses crude chopping, is faster but low accuracy of the implied word.

Lemmatization uses vocabulary analysis, is slower but high accuracy of the implied word.

EX: was - Stemming -> wa , Lemmatization -> be

universal - Stemming -> univers , Lemmatization -> universal

## Task2: Build Advanced Preprocessing Function

```python
import re

# Returns the processed text
def preprocess_text(text):
    
    # TASK7: Error Handling
    if not isinstance(text,str) or text.strip()=="":
        return ""

    # 6. Remove URLs and email like patterns
    text = re.sub(r'http\S+|www\S+', '', text)
    text = re.sub(r'\S+@\S+', '', text)

    # 4. Convert text to lowercase
    text = text.lower()

    # 1. Remove numbers
    text = re.sub(r'\d+', '', text)
    
    # 2. Remove extra spaces
    text = re.sub(r'\s+', ' ', text)
    
    # 3. Handle Repeated characters
    text = re.sub(r'(.)\1{2,}', r'\1', text)

    text = text.strip()

    # 5. Remove very short tokens
    tokens = text.split()
    text = [word for word in tokens if len(word) > 2 or word in ["no", "not"]]

    return " ".join(text)
```

## Task3 : Stress Testing

```python
sample_sentences = [
        "Get 100% FREE access now!!!",
        "I absolutely looooved this product 😍😍",
        "Worst service ever... 0/10",
        "Call me at 9876543210",
        "This is THE best course!!!",
        "Visit https://openai.com now!",
        "Nooooo this is baaad!!!",
        "OK OK OK I got it",
        "Win $$$ now!!! Limited offer!!!",
        "I am not happy with this"
    ]

for sentence in sample_sentences:
    clean_text = preprocess_text(sentence)
    print("-"*30)
    print(f"Original text: {sentence}")
    print(f"Cleaned tokens: {clean_text.split()}")
    print(f"Cleaned sentence: {clean_text}")
```

## Task 4: Token Analytics

```python
count = 1
for sentence in sample_sentences:
    clean_text = preprocess_text(sentence)
    tokens = clean_text.split()
    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    if total_tokens > 0:
        average_token_length = sum(len(word) for word in tokens) / total_tokens
    else:
        average_token_length = 0
    
    print("-"*30)
    print(f"Sentence {count}: {sentence}")
    print(f"Processed sentence: {clean_text}")
    print(f"Total number of tokens: {total_tokens}")
    print(f"Number of unique tokens: {unique_tokens}")
    print(f"Average token length: {average_token_length:.2f}")
    count += 1
```

Analysis questions:

- Sentence 8 had the most noise as after processing, the total tokens reduced to 1 compared to the 6 that can be seen in the original

2) Which sentence retained the most meaningful tokens after cleaning?
- In terms of being accurate and conveying the the best meaning, Sentence 2 is by far the best one. It perfectly captures the expressed meaning.

## Task 5: Frequency Analysis

```python
from collections import Counter

all_tokens = []

for sentence in sample_sentences:
    clean_text = preprocess_text(sentence)
    tokens = clean_text.split()
    all_tokens.extend(tokens)

frequency = Counter(all_tokens)
most_frequent = frequency.most_common(10)
print(f"Top 10 most frequent words: {most_frequent}")

least_frequent = frequency.most_common()[-5:]
print(f"Top 5 least frequent words: {least_frequent}")
```

## Task 6: Build Full Pipeline

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []
    for text in text_list:
        clean_text = preprocess_text(text)
        tokens = clean_text.split()
        all_tokens.extend(tokens)
        clean_sentences.append(clean_text)
    
    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }

full_pipeline(sample_sentences)
```
